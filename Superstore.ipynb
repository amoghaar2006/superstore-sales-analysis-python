```python
print("Converting the Order Date into datetime")
df["Order Date"]=pd.to_datetime(df["Order Date"])

print("Converting the Ship Date into datetime")
df["Ship Date"]=pd.to_datetime(df["Ship Date"])

df["Order Year"]=df["Order Date"].dt.year
df["Ship Year"]=df["Ship Date"].dt.year

df["Order Month"]=df["Order Date"].dt.month_name()
df["Ship Month"]=df["Ship Date"].dt.month_name()
```

```python
print(df[["Order Date","Ship Date","Order Year","Ship Year","Order Month","Ship Month"]].head())
```

```python
print("The Number of Sales in each year")
Sales_Report=df.groupby("Order Year")["Sales"].sum()
# df["Order Year"].value_counts()
Year=Sales_Report.index
Sales=Sales_Report.values

print("LIne Chart representing the total sales for each year")

plt.plot(Year,Sales, color="Blue", marker=".", label="Sales")
plt.xlabel("Year")
plt.ylabel("Total No of Sales")
plt.title("Year v/s Total No of Sales")
plt.tight_layout()
plt.grid(color="red", linewidth="1", linestyle=":")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
print("Category-wise Sales")
Category_Sales=df.groupby("Category")["Sales"].sum()

Category=Category_Sales.index
Sales=Category_Sales.values

plt.bar(Category,Sales,color="blue", label="Sales")
plt.xlabel("Category")
plt.ylabel("No of Sales")
plt.title("Category v/s Sales")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("Region-wise Sales Analysis")
Region_Sales=df.groupby("Region")["Sales"].sum()

Region=Region_Sales.index
Sales=Region_Sales.values

plt.title("Region v/s Total Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.bar(Region,Sales, color="orange", label="Sales")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("Top 10 States by Sales")

State_Sales=df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

State=State_Sales.index
Sales=State_Sales.values

plt.title("State v/s Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.barh(State,Sales,color="green",label="Sales")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
print("Category wise Profit Analysis")
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("Sample - Superstore.csv", encoding="latin1")

Profit_category=df.groupby("Category")["Profit"].sum()

Category=Profit_category.index
Profit=Profit_category.values

plt.bar(Category,Profit, color="pink", label="Profit")
plt.xlabel("Category")
plt.ylabel("Total Profit per Category")
plt.title("Category wise Profit Analysis")


plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Top 10 Products by Sales
import pandas as pd
df=pd.read_csv("Sample - Superstore.csv", encoding="latin1")
import matplotlib.pyplot as plt

Product_Sales=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

Product=Product_Sales.index
Sales=Product_Sales.values

plt.barh(Product, Sales, color="blue", label="Total Sales")
plt.title("Top 10 Products by Sales")
plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.legend()
# plt.tight_layout()
plt.show()
```

```python

Category_percentage_sales=df.groupby("Category")["Sales"].sum()
print(Category_percentage_sales)

Category=Category_percentage_sales.index
Total_Sales=Category_percentage_sales.values

plt.pie(Total_Sales, labels=Category, autopct="%1.1f%%", colors=["red","green","blue"])
plt.title("Pie Chart showing Sales Distribution by Category")
plt.tight_layout()
plt.show()
```

```python
#Sales vs Profit

Sales=df["Sales"]
Profit=df["Profit"]

plt.scatter(Sales,Profit, color="red", label="Orders" ,marker="o")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Scatter Plot showing Sales v/s Profit")
plt.legend()
plt.grid(color="blue",linewidth=1, linestyle=":")
plt.tight_layout()
plt.show()
```

1. Technology generated the highest sales among all categories.
2. Office Supplies had the highest number of orders but not the highest sales.

4. California contributed the highest sales among all states.
5. A few products generated exceptionally high revenue compared to the rest.
